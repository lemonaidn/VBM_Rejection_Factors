```python
party_df.rename(columns = {'ballot_county':'County',
                          'DemTotal':'Dem_Ballots_Cast',
                          'OtherTotal':'Other_Ballots_Cast',
                          'RepubTotal':'Repub_Ballots_Cast',
                          'UnaffTotal':'Unaff_Ballots_Cast',
                          'Democratic':'Dem_Ballots_Rejected',
                          'Other':'Other_Ballots_Rejected',
                          'Republican':'Repub_Ballots_Rejected',
                           'Unaffiliated':'Unaff_Ballots_Rejected',
                          'Dem_per_totalCountyDem%':'Percent_Dem_Ballots_Rejected',
                          'Repub_per_totalCountyRep%':'Percent_Repub_Ballots_Rejected',
                          'Other_per_totalCountyOther%':'Percent_Other_Ballots_Rejected',
                          'Unaff_per_totalCountyUnaff%':'Percent_Unaff_Ballots_Rejected'
                          }, inplace=True)
```

```python
party_df
```

```python
#Let's add columns for the % each party is of total ballots

county_df['Percent_Ballots_Cast_Dem'] = (party_df['Dem_Ballots_Cast'] / county_df['ballots_cast'])*100
county_df['Percent_Ballots_Cast_Repub'] = (party_df['Repub_Ballots_Cast'] / county_df['ballots_cast'])*100
county_df['Percent_Ballots_Cast_Unaff'] = (party_df['Unaff_Ballots_Cast'] / county_df['ballots_cast'])*100
county_df['Percent_Ballots_Cast_Other'] = (party_df['Other_Ballots_Cast'] / county_df['ballots_cast'])*100
```

```python
county_df
```

```python
merged_df = pd.merge(county_df, party_df)
merged_df
```

```python
merged_df = pd.merge(merged_df, demo_df)
merged_df
```
