```python
#Overall percent of rejections
((vbm.received_rejReason.value_counts().sum()/vbm.voter_id.count())*100).round(2)
```

```python
#CALCULATING the number of rejected per county
vbm1=pd.DataFrame(vbm.groupby(['current_county','received_rejReason'])['voter_lastName'].size()).reset_index()
vbm1.columns = ['county','rejReason','rejCount']
(vbm1.groupby('county')['rejCount'].sum()).sort_values(ascending=False)
```

```python
vbm['current_county'].value_counts()
```

```python
vbm1.groupby('county')['rejCount'].sum()
```

```python
#The ratio of rejected per county
(((vbm1.groupby('county')['rejCount'].sum())/vbm['current_county'].value_counts())*100).\
sort_values(ascending=False)
```

```python
#The number of rejections per party
vbm2=pd.DataFrame(vbm.groupby(['voter_party','received_rejReason'])['voter_lastName'].size()).reset_index()
vbm2.columns = ['voter_party','rejReason','rejCount']#.argmax()
vbm2.groupby('voter_party')['rejCount'].sum().sort_values(ascending=False)
```

```python
vbm['voter_party'].value_counts()
```

```python
#The ratio of rejections per party
((vbm2.groupby('voter_party')['rejCount'].sum().sort_values(ascending=False)/vbm.voter_party.value_counts())*100).\
sort_values(ascending=False)
```
